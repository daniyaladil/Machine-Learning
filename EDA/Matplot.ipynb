```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("sharma-kohli.csv")
df
```

```python
plt.plot(df["index"], df["RG Sharma"], label="Rohit", color='c')
plt.plot(df["index"], df["V Kohli"], label="Virat", color='g')

plt.title("Virat vs Rohit IPL Record")
plt.xlabel("Season")        # x is season
plt.ylabel("Runs Scored")   # y is runs
plt.legend()
plt.grid()
plt.show()
```
